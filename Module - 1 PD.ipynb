```python
import os
import pydicom as dicom
import os
import cv2
import PIL
path1 = r'C:\Final Year Project\dataset\T2 AXIAL PD'
```

```python
for path,subdir,dir in os.walk(path1):
    
    for i in dir:
        if i.endswith("dcm"):
            print(os.path.join(path, i))
```

```python
count=0
for path,subdir,dir in os.walk(path1):
    
    for i in dir:
        jpg_folder_path = r"C:\Final Year Project\dataset\T2 AXIAL PD"
        if i.endswith("dcm"):
            
            count +=1
            cnt=str(count)
             # optional
            # make it True if you want in PNG format
            PNG = False
           
            # Specify the output jpg/png folder path
            
            os.chdir(jpg_folder_path)
            
            ds = dicom.dcmread(os.path.join(path, i))
            pixel_array_numpy = ds.pixel_array
           
            i = i.replace('.dcm', '')
            cv2.imwrite(cnt+'.jpg', pixel_array_numpy)
            if count % 50 == 0:
                print('{} image converted'.format(count))
        
```
